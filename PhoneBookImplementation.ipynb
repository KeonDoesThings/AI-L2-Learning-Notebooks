```python
contacts = {}
def addContact():
  contactInfo = []
  contactName = input("What is the name of your new contact?")
  contactPNum = input("What is the phone number of " + contactName + "?")
  contactAdr = input("What is the address of " + contactName + "?")
  contactInfo.append(contactPNum)
  contactInfo.append(contactAdr)
  contacts[contactName] = contactInfo
  mainLoop()

def removeContact():
  if not contacts == False:
    removalChoice = input("Which contact would you like to remove?")
    del contacts[removalChoice]
    print(removalChoice + " has been removed from your phone book!")
    mainLoop()
  else:
    print("You have no contacts to remove!")
def viewContacts():
  if not contacts == False:
    print("Contacts:")
    for x in contacts.keys():
      print(f"{x}")
    mainLoop()
  else:
    print("You have no contacts to view!")

def searchContact():
  searchChoice = input("What contact would you like to search for?")
  if not contacts == False:
    try:
      print(searchChoice + "'s phone number and address is:\n"+contacts[searchChoice][0]+", "+contacts[searchChoice][1])
      mainLoop()
    except:
      print("That contact does not exist in your phone book! Please enter a valid contact.")
      searchContact()
  else:
    print("You have no contacts to view!")

def editContact():
  editContChoice = input("Which contact would you like to edit? (use proper capitalization!)")
  if editContChoice in contacts.keys():
    editChoice = input("Press:\n1: To edit name\n2: To edit phone number\n3: To edit address\n")
    if editChoice == "1":
      newName = input("What would you like to change the name of the contact to?")
      contacts[newName] = contacts.pop(editContChoice)
    elif editChoice == "2":
      newNum = input("What would you like to change the phone number of the contact to?")
      contacts[editContChoice][0] = newNum
    elif editChoice == "3":
      newAdr = input("What would you like to change the address of the contact to?")
      contacts[editContChoice][1] = newAdr
    mainLoop()
  else:
    print("Please enter a valid contact name!")
    editContact()

def mainLoop():
  print("")
  functionChoice = input("What would you like to do?\nPress: \n1: Add a contact\n2: Remove a contact\n3: View contacts\n4: Search a contact\n5: Edit a contact\n")
  if functionChoice == "1":
    print("")
    addContact()
  elif functionChoice == "2":
    print("")
    removeContact()
  elif functionChoice == "3":
    print("")
    viewContacts()
  elif functionChoice == "4":
    print("")
    searchContact()
  elif functionChoice == "5":
    print("")
    editContact()
  else:
    print("Please enter a valid option.")
    mainLoop()

mainLoop()
```
